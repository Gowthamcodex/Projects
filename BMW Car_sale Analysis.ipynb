Interpretation (Insights)

From this box plot, you can understand:
The median price (center line) shows the typical BMW price in your dataset.
The length of the box indicates how spread out the prices are — a longer box means more variation between car prices.

If the median line is not centered in the box, it indicates the distribution is skewed — either more lower-priced or higher-priced vehicles.

A box plot helps visualize the spread, center, and outliers of numerical data — in this case, how BMW prices vary.
The box in the middle represents the interquartile range (IQR), which contains the middle 50% of the data.
The line inside the box is the median price.
The "whiskers" extend from the box to the minimum and maximum values within 1.5 times the IQR from the box.
Any points outside the whiskers are considered outliers.

````python
# Outer ring (FuelType)
outer = df['Fuel_Type'].value_counts()
# Inner ring (Transmission)
inner = df['Transmission'].value_counts()
fig, ax = plt.subplots(figsize=(8,8))
ax.pie(
    outer,
    radius=1,
    labels=outer.index,
    autopct='%1.1f%%',
    colors=plt.cm.Reds(np.linspace(0.3, 1, len(outer)))
)
# Inner ring (Blue shades) – changed color area
ax.pie(
    inner,
    radius=0.7,
    labels=inner.index,
    labeldistance=0.6,
    colors=plt.cm.Blues(np.linspace(0.4, 0.9, len(inner))))
plt.title("Fuel Type vs Transmission Distribution",
          fontsize=14, fontweight='bold', color='darkred')
plt.show()
````
